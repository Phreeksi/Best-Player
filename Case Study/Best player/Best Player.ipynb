```python
team_player = ['Player', 'Point_avg', 'Opp_points_avg', 'avg_win_percentage', 'Number of Games']
team_stat = pd.DataFrame(columns= team_player)
team_stat
```

```python
def winpercentage(dataframe):
    x = dataframe[dataframe['Win/ Loss'] == 'W']
    y = len(x['Win/ Loss'])
    z = (y / len(dataframe)) * 100
    return(z)

def num_games(dataframe):
    x = len(dataframe['Win/ Loss'])
    return(x)


print(num_games(James))
```
